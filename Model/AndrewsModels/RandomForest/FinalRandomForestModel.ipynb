```python
#Training data with seasons 2002 - 2018 every game
train_data = pd.read_csv(r"C:\Users\comic\NFL-Predictor\Data\Train-Data\Compiled\testNOPERCENTAGES.csv")
train_data.head()
#Here is preview of first 5 games 
#Note if outcome is 1 that means team won the game
```

```python
#Testing data with 2019 season
test_data = pd.read_csv(r"C:\Users\comic\NFL-Predictor\Data\Test-Data\Matchups\Matchup2019.csv")
test_data.head()
#Here is a preview of first 5 games 
#Note if outcome is 1 that means team won the game
```

```python
from sklearn.ensemble import RandomForestClassifier


y = train_data["T1 Outcome"]#Want to train data on the first team's outcome column 


features = ["T1 Location", "T2 Location", "T1 Avg Points", "T2 Avg Points", "T1 Avg Yards","T2 Avg Yards", "T1 NPY/A", "T2 NPY/A", "T1 NRY/A", "T2 NRY/A", "T1 Avg Points Def", "T2 Avg Points Def", "T1 Avg Yards Def", "T2 Avg Yards Def", "T1 NPY/A Def", "T2 NPY/A Def","T1 NRY/A Def", "T2 NRY/A Def", "T1 4th Def", "T2 4th Def", "T1 3rd", "T2 3rd", "T1 4th", "T2 4th", "T1 3rd Def", "T2 3rd Def"]
#Here's the features I used to first achieve over 70% prediction accuracy 

x = pd.get_dummies(train_data[features])
x_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators=1000, max_depth=10, random_state=1, min_samples_leaf=25,n_jobs=5, oob_score=True)#Subject to change these parameters, but these seem to work pretty well
model.fit(x, y)
predictions = model.predict(x_test)

output = pd.DataFrame({'T1': test_data.T1, 'T1 Outcome': predictions})
output.to_csv('FinalModel.csv', index=False)#Put model into a .csv file 
```
