```python
from sklearn.ensemble import RandomForestClassifier


y = train_data["T1 Outcome"]

#uncomment below when data is good
features = ["T1 Avg Points", "T2 Avg Points", "T1 Avg Yards","T2 Avg Yards", "T1 NPY/A", "T2 NPY/A", "T1 NRY/A", "T2 NRY/A"]
#Figure out which part of rest is throwing it off , "T1 3rd%", "T2 3rd%", "T1 4th%", "T2 4th%","T1 Avg Points Def", "T2 Avg Points Def", "T1 Avg Yards Def", "T2 Avg Yards Def", "T1 NPY/A Def", "T2 NPY/A Def", "T1 NRY/A Def", "T2 NRY/A Def", "T1 3rd% Def", "T2 3rd% Def", "T1 4th% Def", "T2 4th% Def"
x = pd.get_dummies(train_data[features])
x_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(x, y)
predictions = model.predict(x_test)

output = pd.DataFrame({'T1': test_data.T1, 'T1 Outcome': predictions})
```
