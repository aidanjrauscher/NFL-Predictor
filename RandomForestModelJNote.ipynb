```python
train_data = pd.read_csv(r"C:\Users\comic\NFL-Predictor\NFL-Predictor\Data\Train-Data\Games\Games1995.csv")
train_data.head()
```

```python
test_data = pd.read_csv(r"C:\Users\comic\NFL-Predictor\NFL-Predictor\Data\Train-Data\Games\Games1996.csv")
test_data.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

y = train_data["Winner"]

features = ["PtsW", "PtsL"]
x = pd.get_dummies(train_data[features])
x_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(x, y)
predictions = model.predict(x_test)

output = pd.DataFrame({'Winner': test_data.Winner, 'Loser': predictions})
```
